## 9.1.1 - Valor Presente (VP)
    - formula: VP = FV / (1 + i)^n

```python
def vp(vf, i, n):
    return vf / (1 + i)**n

print(vp(100, 0.1, 1))
```

## 9.1.2 - Valor Presente Liquido (VPL)
- formula: VPL = VP1 + VP2 + VP3 + ... + VPn

```python
import numpy as np
import numpy_financial as npf

vinicial = -155000
esaidas = 55000
fc = np.array([vinicial, esaidas, esaidas, esaidas, esaidas, esaidas, esaidas])

investimento_a_valor_presente = npf.npv(0.17, fc)
print(investimento_a_valor_presente)
```

## 9.1.3 - Taxa Interna de Retorno (TIR)
VPL = FC1 / (1 + i)^1 + FC2 / (1 + i)^2 + ... + FCn / (1 + i)^n

Para calcular a TIR, precisamos encontrar o valor de i que faz com que o VPL seja igual a zero.
- pq zerando o vpl so vai restar de incoginta a taxa de retorno

```python
# Calculando a TIR do VPL anterior
import numpy as np
import numpy_financial as npf

vinicial = -155000
esaidas = 55000
fc = np.array([vinicial, esaidas, esaidas, esaidas, esaidas, esaidas, esaidas])

tir = npf.irr(fc)

print(f"{round(100 * tir, 2)}%")
```

```python
# Grafico de TIR
import numpy as np
import numpy_financial as npf
import matplotlib.pyplot as fig

vinicial = -155000
esaidas = 55000
fc = np.array([vinicial, esaidas, esaidas, esaidas, esaidas, esaidas, esaidas])

tir = npf.irr(fc)

# vou criar um array de taxas de retorno de 0% a 35% com 200 pontos entre eles
eixox = np.linspace(0, 0.35, 200)
# Calculando o VPL para cada taxa de retorno, um vetor dos 200 caso de VPLs
vpl = np.array([npf.npv(x, fc) for x in eixox])

# Fazendo o eixo X e Y do VPL
fig.plot(eixox, vpl, label="VPL", linewidth=2)
fig.plot(tir, 0, 'ro', label="TIR", markersize=15)
fig.title("Taxa Interna de Retorno")
fig.xlabel("Juros considerado (i)")
fig.ylabel("VPL")
fig.grid()
```
